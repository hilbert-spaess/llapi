```python
import os
```

```python
COURSE_NO = "5"
COURSE_DIRECTORY = "/home/jc2062/llapi/courses"
PARAS_FILE= os.path.join(COURSE_DIRECTORY, COURSE_NO, "paras_raw")
```

```python
with open(PARAS_FILE, 'r') as parasfile:
    
    lines = parasfile.readlines()
    
    lines  = [x.strip() for x in lines if x.strip()]
    
    lines = [lines[i] for i in range(1,len(lines)) if lines[i-1][:9]=="Passage 4" and lines[i][-1] in [".", "!", "?"]]
    
    lines = [x for x in lines if len([y for y in x if y in [".", "?", "!"]]) > 1]
```

```python
lines
```

```python
with open("all_paras.txt", 'a') as outfile:
    for line in lines:
        outfile.write(line + "\n\n")
```
