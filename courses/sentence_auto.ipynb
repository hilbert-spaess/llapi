```python
# use pyautogui to automate.
# setup: firefox open in two adjacent windows

import pyautogui
import time
import os
```

```python
COURSE="./3_ETHAN" # course directory
LAPTOP="XPS" # which laptop the code is being run on: XPS, ASUS
VOCABFILENAME = os.path.join(COURSE, "vocab.txt")
print(VOCABFILENAME)
```

```python
with open(VOCABFILENAME, 'r') as vocabfile:
    words = [x.strip() for x in vocabfile.readlines()]
```

```python
print(words)
```

```python
# refresh
def refresh():
    if LAPTOP=="XPS":
        pyautogui.click(340, 980)
```

```python
# undo
def undo():
    if LAPTOP=="XPS":
        pyautogui.click(250, 980)
```

```python
# edit
def edit():
    if LAPTOP=="XPS":
        pyautogui.click(190, 980)
```

```python
def confirm_edit():
    if LAPTOP=="XPS":
        pyautogui.click(550, 830)
```

```python
def go():
    if LAPTOP=="XPS":
        pyautogui.click(550, 1020)
```

```python
def copy_progress():
    if LAPTOP=="XPS":
        pyautogui.click(400, 700)
    pyautogui.keyDown('ctrl')
    pyautogui.press('a')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('c')
    pyautogui.keyUp('ctrl')
    if LAPTOP=="XPS":
        pyautogui.click(1600, 700)
    pyautogui.keyDown('ctrl')
    pyautogui.press('end')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('v')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('s')
    pyautogui.keyUp('ctrl')
```

```python
def iteration(prompt):
    
    # undo
    
    undo()
    
    time.sleep(1)
    
    # edit
    
    edit()
    
    time.sleep(1)
    
    # delete everything so far
    
    edit()
    time.sleep(1)
    
    pyautogui.keyDown('ctrl')
    pyautogui.press('a')
    pyautogui.keyUp('ctrl')
    pyautogui.press('backspace')
    
    # write in a new prompt
    
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.write(prompt)
    """
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    """
    
    time.sleep(5)
    
    # generate
    
    confirm_edit()
    
    time.sleep(5)
    
    go()
    
    time.sleep(10)
    
    copy_progress()
    
    
    for i in range(10):
        print(i)
        refresh()
        time.sleep(10)
        copy_progress()
    
    # 
    
    # reload and copy a certain number of times
```
