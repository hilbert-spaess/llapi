```python
# refresh
def refresh():
    if LAPTOP=="XPS":
        pyautogui.click(270, 1020)
    if LAPTOP=="ASUS":
        pyautogui.click(340, 650)
```

```python
# undo
def undo():
    if LAPTOP=="XPS":
        pyautogui.click(200, 1020)
    if LAPTOP=="ASUS":
        pyautogui.click(250, 650)
```

```python
# edit
def edit():
    if LAPTOP=="XPS":
        pyautogui.click(170, 1020)
    if LAPTOP=="ASUS":
        pyautogui.click(190, 650)
```

```python
def confirm_edit():
    if LAPTOP=="XPS":
        pyautogui.click(550, 830)
    if LAPTOP=="ASUS":
        pyautogui.click(590, 500)
```

```python
def go():
    if LAPTOP=="XPS":
        pyautogui.click(550, 1020)
    if LAPTOP=="ASUS":
        pyautogui.click(590, 680)
```

```python
def copy_progress():
    if LAPTOP=="XPS":
        pyautogui.click(400, 700)
    if LAPTOP=="ASUS":
        pyautogui.click(400, 400)
    pyautogui.keyDown('ctrl')
    pyautogui.press('a')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('c')
    pyautogui.keyUp('ctrl')
    if LAPTOP=="XPS":
        pyautogui.click(1600, 700)
    if LAPTOP=="ASUS":
        pyautogui.click(800, 600)
    pyautogui.keyDown('ctrl')
    pyautogui.press('end')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('v')
    pyautogui.keyUp('ctrl')
    pyautogui.keyDown('ctrl')
    pyautogui.press('s')
    pyautogui.keyUp('ctrl')
```

```python
pyautogui.click(480, 900)
```

```python
def iteration(prompt):
    
    # undo
    
    undo()
    
    time.sleep(1)
    
    # edit
    
    # delete everything so far
    
    edit()
    time.sleep(1)
    
    pyautogui.click(480, 900)
    
    pyautogui.keyDown('ctrl')
    pyautogui.press('a')
    pyautogui.keyUp('ctrl')
    pyautogui.press('backspace')
    
    # write in a new prompt
    
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.write(prompt)
    """
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    pyautogui.keyDown('shift')
    pyautogui.press('enter')
    pyautogui.keyUp('shift')
    """
    
    time.sleep(5)
    
    # generate
    
    confirm_edit()
    
    time.sleep(5)
    
    go()
    
    time.sleep(10)
    
    copy_progress()
    
    
    for i in range(70):
        print(i)
        refresh()
        time.sleep(10)
        copy_progress()
    
    # 
    
    # reload and copy a certain number of times
```
